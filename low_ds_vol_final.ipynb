```python
import pandas as pd
file=r'T:\EquityPd\1-Projects\Janus Henderson\Low Volatility\1-Selection\price_valuepasted2.xlsx'
df=pd.read_excel(file, parse_dates=True, index_col=0)
file2=r'T:\EquityPd\1-Projects\Janus Henderson\Low Volatility\1-Selection\universe.xlsx'
universe=pd.read_excel(file2, parse_dates=True, index_col=0)
```

```python
#calendar
period='Q-JAN' #choose 'Y', 'Q', 'M'

calendar=pd.DataFrame()
calendar['rebalancing_date']=df.index
calendar['252days_ago']=df.index
calendar['252days_ago']=calendar['252days_ago'].shift(+251)

calendar.index=calendar.rebalancing_date
calendar_Q=calendar.resample(period).last().asfreq(period)
calendar_Q_adj=calendar_Q.loc["2000-1-31":,:]
calendar_Q_adj
```

```python

#q=df.resample(period).last().asfreq(period)
#q['t-1']=q.index.shift(shift)
data=pd.DataFrame()

# Time series
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
%matplotlib inline
import datetime 


def get_price(period,shift):
    data=pd.DataFrame()
    dt=pd.DataFrame
    for i in range(len(calendar_Q_adj.index)):
        dt=df.pct_change()
        dt=dt[dt<0]
        dt=dt.fillna(0)
        
        date1=calendar_Q_adj.iloc[i,0]
        date2=calendar_Q_adj.iloc[i,1]
        dt=dt[(dt.index<=date1) & (dt.index>=date2)]
        dt=dt.std()*252**(0.5)

        data[calendar_Q_adj.iloc[i,0]]=dt
    return data

def universe_screen(data):
    screen=universe>0
    screen=screen.astype('int32')
    data=data*screen
    return data

#def trading_day_check(shift, data):
#    price_check=(data.shift(-shift,axis=1)!=0) #0 if there is no trading price 1 year ago (problem: delisted within 1 year, ie 1 month ago)
#    price_check.astype('int32')
#    trading_day_check=data*price_check
#    return trading_day_check

def trading_day_check(data):
    trading=pd.DataFrame()
    for i in range(len(calendar_Q_adj.index)):
        date1=calendar_Q_adj.iloc[i,0]
        date2=calendar_Q_adj.iloc[i,1]
        #sample=df[(df.index<=date1) & (df.index>=date2)]!=0
        #sample=sample.astype('int32').sum(axis=0)==252
        sample=df[(df.index<=date1) & (df.index>=date2)].isnull()
        sample=sample.astype('int32').sum(axis=0)==0
        sample=sample.astype('int32')
        trading[calendar_Q_adj.iloc[i,0]]=sample

    trading_day_check=data*trading
    return trading_day_check

def inverse_vol(data):
    n=1/data
    n=n.replace(np.inf, np.nan)
    return n

def adj_inverse_vol(n):
    n= inverse_vol(data)
    n_rank=n.rank(axis=0, ascending=0)<=100
    n_rank=n_rank.astype('int32')
    inverse = n_rank*n
    inverse=inverse.fillna(0)
    return inverse

def initial_w(n):
    w=n/n.sum(axis=0)
    #w=w.transpose()
    return w

def adj_wgt(wgt_limit):
    w1=(w<wgt_limit)&(w>0)
    w1=w1.astype('int32')
    w2=pd.DataFrame()
    w2=w
    w2[w2>wgt_limit]=wgt_limit

    w3=1-w2.sum(axis=0)
    w4=w1*w

    #divide matrix by 1 column
    w5=w4.divide(w4.sum(axis=0), axis=1)
    w6=w5.multiply(w3, axis=1)
    weight=w6+w2
    return weight

wgt_limit=0.03
period='Q-JAN' #choose 'Y', 'Q', 'M'
shift=-4 #-4 if 'Q' and the reference date is 1yr ago

data = get_price(period,shift)
data=universe_screen(data)
#data=trading_day_check(shift,data)
data=trading_day_check(data)
n= inverse_vol(data)
inverse=adj_inverse_vol(n)
w=initial_w(inverse)
result=adj_wgt(wgt_limit)

result


```
